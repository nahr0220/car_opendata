```python
# new_df의 월별 카운트 계산
new_monthly_count = new_df['월'].value_counts().sort_index()
new_monthly_count_df = new_monthly_count.reset_index()
new_monthly_count_df.columns = ['월', 'new_카운트']

# used_df의 월별 카운트 계산
used_monthly_count = used_df['월'].value_counts().sort_index()
used_monthly_count_df = used_monthly_count.reset_index()
used_monthly_count_df.columns = ['월', 'used_카운트']

# 두 데이터프레임 병합 (outer join으로 누락된 월도 포함)
combined_monthly_count_df = pd.merge(new_monthly_count_df, used_monthly_count_df, on='월', how='outer')

# NaN 값을 0으로 채우기
combined_monthly_count_df.fillna(0, inplace=True)

# 데이터 타입 변환 (카운트는 정수형으로 변환)
combined_monthly_count_df['new_카운트'] = combined_monthly_count_df['new_카운트'].astype(int)
combined_monthly_count_df['used_카운트'] = combined_monthly_count_df['used_카운트'].astype(int)

combined_monthly_count_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

# Windows에서 한글 폰트 설정 (맑은 고딕)
rc('font', family='Malgun Gothic')

# 한글 깨짐 방지 (음수 기호 처리)
plt.rcParams['axes.unicode_minus'] = False

# 월별 전체 합계 데이터 계산
combined_monthly_count_df['전체_합계'] = combined_monthly_count_df['new_카운트'] + combined_monthly_count_df['used_카운트']

# 시각화 (공통 축 사용)
fig, ax = plt.subplots(figsize=(10, 6))

# 막대 그래프
x = combined_monthly_count_df['월']
bar_width = 0.35

# 신규, 이전 데이터를 막대로 표시
ax.bar(x, combined_monthly_count_df['new_카운트'], width=bar_width, label='신규', align='center')
ax.bar(x, combined_monthly_count_df['used_카운트'], width=bar_width, label='이전', align='edge')

# 꺾은선 그래프 (전체 합계)
ax.plot(x, combined_monthly_count_df['전체_합계'], color='black', marker='o', label='전체 합계', linewidth=2)

# 축 레이블 및 제목 설정
ax.set_ylabel('개수')
ax.set_title('전체 등록 현황', fontsize=18, weight='bold')
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.legend(loc='best')

# 격자 추가
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# new_df의 회원등록번호별 카운트 계산
new_age_count = new_df['회원등록번호'].value_counts(dropna=False).sort_index()
new_age_count_df = new_age_count.reset_index()
new_age_count_df.columns = ['회원등록번호', 'new_카운트']

# used_df의 회원등록번호별 카운트 계산
used_age_count = used_df['회원등록번호'].value_counts(dropna=False).sort_index()
used_age_count_df = used_age_count.reset_index()
used_age_count_df.columns = ['회원등록번호', 'used_카운트']

# 두 데이터프레임 병합 (outer join으로 누락된 회원등록번호도 포함)
combined_age_count_df = pd.merge(new_age_count_df, used_age_count_df, on='회원등록번호', how='outer')

# NaN 값을 0으로 채우기
combined_age_count_df.fillna('데이터없음', inplace=True)

# 데이터 타입 변환 (카운트는 정수형으로 변환)
combined_age_count_df['new_카운트'] = combined_age_count_df['new_카운트'].astype(int)
combined_age_count_df['used_카운트'] = combined_age_count_df['used_카운트'].astype(int)

combined_age_count_df
```

```python
import numpy as np

# 데이터 준비
combined_age_count_df = combined_age_count_df.sort_values(by='회원등록번호')
labels = combined_age_count_df['회원등록번호'].values
new_data = combined_age_count_df['new_카운트'].values
used_data = combined_age_count_df['used_카운트'].values

# 막대 위치 설정
x = np.arange(2)  # 두 개의 막대: 신규, 이전
bar_width = 0.4  # 막대 너비

# 그래프 생성
fig, ax = plt.subplots(figsize=(10, 6))

# 누적 막대 데이터를 위한 초기화
cumulative_new = np.zeros(1)  # 신규 누적값 초기화
cumulative_used = np.zeros(1)  # 이전 누적값 초기화

# 색상 고정 설정
colors = plt.cm.tab10(np.linspace(0, 1, len(labels)))  # 회원등록번호별 색상 고정

# '신규' 막대
for i, (label, value) in enumerate(zip(labels, new_data)):
    ax.bar(x[0], value, width=bar_width, color=colors[i], label=f'회원등록번호 {label}' if i == 0 else "", bottom=cumulative_new)
    cumulative_new += value

# '이전' 막대
for i, (label, value) in enumerate(zip(labels, used_data)):
    ax.bar(x[1], value, width=bar_width, color=colors[i], label=f'회원등록번호 {label}' if i == 0 else "", bottom=cumulative_used)
    cumulative_used += value

# 축 레이블 및 제목
ax.set_xticks(x)
ax.set_xticklabels(['신규', '이전'])
ax.set_ylabel('카운트')
ax.set_title('연령대별 등록 현황', fontsize=18, weight='bold')

# 범례 설정 (회원등록번호별)
handles = [plt.Rectangle((0, 0), 1, 1, color=colors[i]) for i in range(len(labels))]
ax.legend(handles, labels, title='회원등록번호', loc='best')

# 격자 추가
ax.grid(axis='y', linestyle='--', alpha=0.7)

# 그래프 출력
plt.tight_layout()
plt.show()
```

```python
# new_df의 월별 카운트 계산
new_gender_count = new_df['성별'].value_counts().sort_index()
new_gender_count_df = new_gender_count.reset_index()
new_gender_count_df.columns = ['성별', 'new_카운트']

# used_df의 월별 카운트 계산
used_gender_count = used_df['성별'].value_counts().sort_index()
used_gender_count_df = used_gender_count.reset_index()
used_gender_count_df.columns = ['성별', 'used_카운트']

# 두 데이터프레임 병합 (outer join으로 누락된 성별 포함)
combined_gender_count_df = pd.merge(new_gender_count_df, used_gender_count_df, on='성별', how='outer')

# NaN 값을 0으로 채우기
combined_gender_count_df.fillna(0, inplace=True)

# 데이터 타입 변환 (카운트는 정수형으로 변환)
combined_gender_count_df['new_카운트'] = combined_gender_count_df['new_카운트'].astype(int)
combined_gender_count_df['used_카운트'] = combined_gender_count_df['used_카운트'].astype(int)

combined_gender_count_df
```

```python
# 'new_비율'과 'used_비율' 계산
combined_gender_count_df['new_비율'] = combined_gender_count_df['new_카운트'] / combined_gender_count_df['new_카운트'].sum()
combined_gender_count_df['used_비율'] = combined_gender_count_df['used_카운트'] / combined_gender_count_df['used_카운트'].sum()

# 신규 데이터와 이전 데이터를 분리
new_data = combined_gender_count_df[['성별', 'new_비율']].rename(columns={'new_비율': '비율'})
new_data['구분'] = '신규'

used_data = combined_gender_count_df[['성별', 'used_비율']].rename(columns={'used_비율': '비율'})
used_data['구분'] = '이전'

# 데이터 합치기
plot_data = pd.concat([new_data, used_data])

# 피벗 테이블로 성별 비율 재구성
pivot_data = plot_data.pivot_table(index='구분', columns='성별', values='비율', fill_value=0)

# 그래프 그리기
fig, ax = plt.subplots(figsize=(8, 6))

bar_width = 0.6
x = range(len(pivot_data))

# 각각의 성별 비율을 누적하여 표현
bottom_values = [0] * len(pivot_data)
for gender in pivot_data.columns:
    ax.bar(x, pivot_data[gender], width=bar_width, label=gender, bottom=bottom_values, alpha=0.7)
    bottom_values = [b + v for b, v in zip(bottom_values, pivot_data[gender])]

# 그래프 꾸미기
ax.set_xticks(x)
ax.set_xticklabels(pivot_data.index)
ax.set_xlabel('구분')
ax.set_ylabel('비율')
ax.set_title('성별 비율 분포', fontsize=18, weight='bold')
ax.legend(title='성별')

plt.tight_layout()
plt.show()
```

```python
# 회원등록번호를 문자열로 변환 후 조건에 따라 "70대 이상"으로 변경
new_df['회원등록번호'] = new_df['회원등록번호'].astype(str)
new_df.loc[new_df['회원등록번호'].str.startswith(('70', '80', '90')), '회원등록번호'] = '70대 이상'

used_df['회원등록번호'] = used_df['회원등록번호'].astype(str)
used_df.loc[used_df['회원등록번호'].str.startswith(('70', '80', '90')), '회원등록번호'] = '70대 이상'

# new_df의 성별과 회원등록번호 기준 그룹화 및 카운트 계산
new_grouped_count = new_df.groupby(['성별', '회원등록번호']).size().reset_index(name='new_카운트')

# NaN 값을 "빈값"으로 표시
new_grouped_count['성별'] = new_grouped_count['성별'].fillna('빈값')
new_grouped_count['회원등록번호'] = new_grouped_count['회원등록번호'].fillna('빈값')

# used_df의 성별과 회원등록번호 기준 그룹화 및 카운트 계산
used_grouped_count = used_df.groupby(['성별', '회원등록번호']).size().reset_index(name='used_카운트')

# NaN 값을 "빈값"으로 표시
used_grouped_count['성별'] = used_grouped_count['성별'].fillna('빈값')
used_grouped_count['회원등록번호'] = used_grouped_count['회원등록번호'].fillna('빈값')

# 두 데이터프레임 병합 (성별과 회원등록번호 기준으로 outer join)
combined_grouped_count = pd.merge(
    new_grouped_count, 
    used_grouped_count, 
    on=['성별', '회원등록번호'], 
    how='outer'
)

# NaN 값을 0으로 채우기
combined_grouped_count.fillna(0, inplace=True)

# 데이터 타입 변환 (카운트는 정수형으로 변환)
combined_grouped_count['new_카운트'] = combined_grouped_count['new_카운트'].astype(int)
combined_grouped_count['used_카운트'] = combined_grouped_count['used_카운트'].astype(int)

combined_grouped_count
```

```python
male_data = combined_grouped_count[combined_grouped_count['성별'] == '여자']
# 회원등록번호별로 new_카운트 비율 계산
male_data['new_비율'] = male_data['new_카운트'] / male_data['new_카운트'].sum()

# 남성 데이터의 new_카운트 비율로 파이 차트 생성
fig, ax = plt.subplots(figsize=(10, 8))

# 파이 차트 그리기
ax.pie(
    male_data['new_비율'], 
    labels=male_data['회원등록번호'], 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=plt.cm.Paired.colors,
    textprops={'fontsize': 14}
)

# 그래프 제목
ax.set_title('여성의 나이대 비율', fontsize=18, weight='bold')

plt.tight_layout()
plt.show()
```
