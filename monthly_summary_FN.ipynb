```python
# 취득금액별로 그룹핑
bins = [0, 10000000, 20000000, 30000000, 40000000, 60000000, 80000000, 100000000, float('inf')]
labels = ['~1천만원', '~2천만원', '~3천만원', '~4천만원', '~6천만원', '~8천만원', '~1억원', '1억원 초과']

df['취득금액_범위'] = pd.cut(df['취득금액'], bins=bins, labels=labels, right=True)
df['취득금액_범위'] = df['취득금액_범위'].cat.add_categories('데이터없음')  # 새로운 범주 추가
df['취득금액_범위'] = df['취득금액_범위'].fillna('데이터없음')  # NaN을 '값없음'으로 변경

df['취득금액_범위']
```

```python
# 공백 정리: 연속된 공백을 하나의 공백으로 변환
df['사용본거지주소명'] = df['사용본거지주소명'].str.replace(r'\s+', ' ', regex=True).str.strip()
df['사용본거지주소명'] = df['사용본거지주소명'].str.strip()
df_split = df['사용본거지주소명'].str.split(' ', expand=True)

# 각 부분을 새로운 열로 추가 (열 이름 지정 가능)
df[['시/도', '구/군', '동']] = df_split.iloc[:, :3]  # 앞 3개의 컬럼만 선택
df['시/도'] = df['시/도'].replace('전라북도', '전북특별자치도')
df['시/도'] = df['시/도'].replace('강원도', '강원특별자치도')
```

```python
def group_fun(df, group_cols):
    """
    그룹화된 데이터 요약 함수
    """
    return df.groupby(group_cols).agg(
        count=('취득금액', 'count'),         # 카운트
        # total_sales=('취득금액', 'sum'),     # 합계
        # average_sales=('취득금액', 'mean')   # 평균
    )


# 월별 + 그룹별 분석
result = group_fun(df, ['월', '성별'])  # 월별 + 성별
result1 = group_fun(df, ['월', '나이'])  # 월별 + 나이별
result2 = group_fun(df, ['월', '시/도'])  # 월별 + 지역별
result3 = group_fun(df, ['월', '성별', '나이'])  # 월별 + 성별 + 나이별
# result4 = group_fun(df, ['월', '차종'])  # 월별 + 차종
result5 = group_fun(df, ['월', '취득금액_범위'])  # 월별 + 취득금액구간별
result6 = group_fun(df, ['월', '주행거리_범위'])  # 월별 + 주행거리구간별
result7 = group_fun(df, ['월', '구/군'])  # 월별 + 지역별
result8 = group_fun(df, ['월', '시/도', '구/군'])  # 월별 + 지역별
```

```python
result3 = df.groupby(["월", "성별", "나이"]).size().reset_index(name="count")
unknown = result3[result3["성별"] == "데이터없음"]
known = result3[result3["성별"] != "데이터없음"]

pivot_df = known.pivot(index=["월", "성별"], columns="나이", values="count").reset_index()
unknown_sum = unknown.groupby("월")["count"].sum().reset_index().rename(columns={"count": "데이터없음"})
pivot_df = pivot_df.merge(unknown_sum, on="월", how="left").fillna(0)

# ✅ 중복 제거: 월별 첫 행만 데이터없음 값 유지
pivot_df["데이터없음"] = pivot_df.groupby("월")["데이터없음"].transform(
    lambda x: [x.iloc[0]] + [0] * (len(x) - 1)
)

pivot_df = pivot_df.sort_values(by=["월", "성별"]).reset_index(drop=True)

# 열 정렬 (선택)
age_order = ['10대', '20대', '30대', '40대', '50대', '60대', '70대', '80대', '90대']
final_columns = ['월', '성별'] + [col for col in age_order if col in pivot_df.columns] + ['데이터없음']
pivot_df = pivot_df[final_columns]

pivot_df
```
