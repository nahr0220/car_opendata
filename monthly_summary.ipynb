```python
df = df[(df['이전등록구분명'] == '매매업자거래이전') & (df['등록상세명'] == '일반소유용')]
df = df[df['회원구분명'] == '개인']
```

```python
# 취득금액별로 그룹핑
bins = [0, 10000000, 20000000, 30000000, 40000000, 60000000, 80000000, 100000000, float('inf')]
labels = ['~1천만원', '~2천만원', '~3천만원', '~4천만원', '~6천만원', '~8천만원', '~1억원', '1억원 초과']

df['취득금액_범위'] = pd.cut(df['취득금액'], bins=bins, labels=labels, right=True)
df['취득금액_범위'] = df['취득금액_범위'].cat.add_categories('데이터없음')  # 새로운 범주 추가
df['취득금액_범위'] = df['취득금액_범위'].fillna('데이터없음')  # NaN을 '값없음'으로 변경

df['취득금액_범위']
```

```python
# 주행거리별로 그룹핑
bins = [0, 10000, 20000, 40000, 60000, 80000, 100000, 120000, 140000, 160000, 180000, 200000, float('inf')]
labels = ['~1만km', '~2만km', '~4만km', '~6만km', '~8만km', '~10만km', '~12만km', '~14만km', '~16만km', '~18만km', '~20만km', '20만km초과']

df['주행거리_범위'] = pd.cut(df['주행거리'], bins=bins, labels=labels, right=True)
df['주행거리_범위'] = df['주행거리_범위'].cat.add_categories('데이터없음')  # 새로운 범주 추가
df['주행거리_범위'] = df['주행거리_범위'].fillna('데이터없음')  # NaN을 '값없음'으로 변경

df['주행거리_범위']
```

```python
# 성별 나이 텍스트 나누기
# df = df.dropna(subset=['회원등록번호']) # 빈값 삭제(총 건)
df[['나이', '성별']] = df['회원등록번호'].str.split(' ', expand=True)
```

```python
df.info()
```

```python
# 성별이 없는 경우 "데이터 없음"으로 채우기
df['성별'] = df['성별'].fillna('한데이터없음')

# 빈 문자열도 처리 (필요 시)
df['성별'] = df['성별'].replace('', '한데이터없음')

# 성별이 없는 경우 "데이터 없음"으로 채우기
df['나이'] = df['나이'].fillna('한데이터없음')

# 빈 문자열도 처리 (필요 시)
df['나이'] = df['나이'].replace('', '한데이터없음')
```

```python
# 공백 정리: 연속된 공백을 하나의 공백으로 변환
df['사용본거지주소명'] = df['사용본거지주소명'].str.replace(r'\s+', ' ', regex=True).str.strip()
df['사용본거지주소명'] = df['사용본거지주소명'].str.strip()
df_split = df['사용본거지주소명'].str.split(' ', expand=True)

# 각 부분을 새로운 열로 추가 (열 이름 지정 가능)
df[['시/도', '구/군', '동']] = df_split.iloc[:, :3]  # 앞 3개의 컬럼만 선택
df
```

```python
def group_fun(df, group_cols):
    """
    그룹화된 데이터 요약 함수
    """
    return df.groupby(group_cols).agg(
        count=('취득금액', 'count'),         # 카운트
        # total_sales=('취득금액', 'sum'),     # 합계
        # average_sales=('취득금액', 'mean')   # 평균
    )


# 월별 + 그룹별 분석
result = group_fun(df, ['월', '성별'])  # 월별 + 성별
result1 = group_fun(df, ['월', '나이'])  # 월별 + 나이별
result2 = group_fun(df, ['월', '시/도'])  # 월별 + 지역별
result7 = group_fun(df, ['월', '구/군'])  # 월별 + 지역별
result8 = group_fun(df, ['월', '시/도', '구/군'])  # 월별 + 지역별
result3 = group_fun(df, ['월', '성별', '나이'])  # 월별 + 성별 + 나이별
result4 = group_fun(df, ['월', '차종'])  # 월별 + 차종
result5 = group_fun(df, ['월', '취득금액_범위'])  # 월별 + 취득금액구간별
result6 = group_fun(df, ['월', '주행거리_범위'])  # 월별 + 주행거리구간별
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

# Windows에서 한글 폰트 설정 (맑은 고딕)
rc('font', family='Malgun Gothic')

# 한글 깨짐 방지 (음수 기호 처리)
plt.rcParams['axes.unicode_minus'] = False

# 그룹화
grouped = df.groupby(['성별', '회원등록번호'])['취득금액'].sum().reset_index()

# 막대 그래프
plt.figure(figsize=(8, 6))
sns.barplot(data=grouped, x='회원등록번호', y='취득금액', hue='성별', palette='pastel')

# 그래프 제목 및 레이블 설정
plt.title('성별과 나이에 따른 취득금액 합계', fontsize=16)
plt.xlabel('나이', fontsize=12)
plt.ylabel('취득금액 합계', fontsize=12)
plt.legend(title='성별', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 그래프 표시
plt.show()
```
